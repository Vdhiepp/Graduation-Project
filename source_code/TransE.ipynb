```python
# Load triple factory
tf = TriplesFactory.from_path("kg_triples.tsv")

# Tạo lại mô hình (same config)
model = TransE(triples_factory=tf, embedding_dim=100)

# Load trọng số từ file
model.load_state_dict(torch.load("trained_transe/model_state.pt", map_location="cpu"))
model = model.eval()
```

```python
# Load lại CSV chứa KG + ảnh
df = pd.read_csv("coco_kg_triplets.csv")
df["subject"] = df["subject"].str.strip().str.lower()
df["relation"] = df["relation"].str.strip().str.lower()
df["object"] = df["object"].str.strip().str.lower()

# Lấy mapping từ triple factory
entity_to_id = tf.entity_to_id
relation_to_id = tf.relation_to_id
```

```python
# Hàm tính embedding + tìm gần nhất
```

```python
import torch.nn.functional as F
import torch

def get_triplet_embedding(h, r, t):
    h_id = entity_to_id.get(h)
    r_id = relation_to_id.get(r)
    t_id = entity_to_id.get(t)

    if h_id is None or r_id is None or t_id is None:
        raise ValueError(f"Không tìm thấy: ({h}, {r}, {t}) trong KG")

    h_emb = model.entity_representations[0](torch.tensor([h_id]))
    r_emb = model.relation_representations[0](torch.tensor([r_id]))
    t_emb = model.entity_representations[0](torch.tensor([t_id]))
    return h_emb, r_emb, t_emb
```

```python
# Hàm tìm triplet gần nhất
```

```python
def search_similar_triplets(query_h, query_r, query_t, top_k=5):
    query_h, query_r = query_h.lower(), query_r.lower()
    try:
        h_emb, r_emb, t_emb = get_triplet_embedding(query_h, query_r, query_t.lower())
    except ValueError as e:
        print(e)
        return []

    query_vec = h_emb + r_emb
    results = []

    for _, row in df.iterrows():
        h, r, t = row["subject"], row["relation"], row["object"]
        try:
            h_id = entity_to_id[h]
            r_id = relation_to_id[r]
            h_emb = model.entity_representations[0](torch.tensor([h_id]))
            r_emb = model.relation_representations[0](torch.tensor([r_id]))
            candidate_vec = h_emb + r_emb
            sim = F.cosine_similarity(query_vec, candidate_vec).item()
            results.append((h, r, t, row["image_id"], row["caption"], sim))
        except KeyError:
            continue

    results.sort(key=lambda x: x[-1], reverse=True)
    return results[:top_k]
```

```python
results = search_similar_triplets("dog", "catch", "frisbee", top_k=5)

for i, (h, r, t, img_id, caption, sim) in enumerate(results, 1):
    print(f"{i}. ({h}, {r}, {t}) | image_id: {img_id} | sim: {sim:.4f}")
    print(f"   ↳ caption: {caption}\n")
```

```python
from transformers import BlipProcessor, BlipForConditionalGeneration
from PIL import Image
import torch

# Load mô hình BLIP
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
model_blip = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

def generate_caption_from_image(image_path):
    image = Image.open(image_path).convert("RGB")
    inputs = processor(images=image, return_tensors="pt")
    out = model_blip.generate(**inputs)
    caption = processor.decode(out[0], skip_special_tokens=True)
    return caption
```

```python
# Dùng BLIP để tạo caption từ ảnh
caption = generate_caption_from_image("E:/Download/val2017/000000014439.jpg")
print("Caption:", caption)
```

```python
import spacy

# Load mô hình tiếng Anh của spaCy
nlp = spacy.load("en_core_web_sm")

def extract_triplets(caption):
    doc = nlp(caption)
    triplets = []

    for sent in doc.sents:
        subject = None
        object_ = None
        verb = None

        for token in sent:
            # Chủ ngữ
            if "subj" in token.dep_:
                subject = token.text
                verb = token.head
            # Tân ngữ
            if "obj" in token.dep_:
                object_ = token.text

        if subject and verb and object_:
            triplets.append((subject.lower(), verb.lemma_.lower(), object_.lower()))
        elif verb and object_:
            triplets.append(("unknown", verb.lemma_.lower(), object_.lower()))
        elif subject and verb:
            triplets.append((subject.lower(), verb.lemma_.lower(), "unknown"))

    return triplets
```

```python
# Trích xuất triplet từ caption
triplets = extract_triplets(caption)
print("Triplets:", triplets)
```

```python
import pandas as pd

# Đường dẫn đến file CSV triplet 
df = pd.read_csv("coco_kg_triplets.csv")


def search_similar_triplets_relaxed(query_h, query_r, query_t, top_k=10):
    query_h, query_r, query_t = query_h.lower(), query_r.lower(), query_t.lower()

    try:
        h_emb, r_emb, t_emb = get_triplet_embedding(query_h, query_r, query_t)
        query_vec = h_emb + r_emb
        strict_mode = True
    except ValueError:
        # Nếu thiếu 1 phần → chuyển sang relaxed mode
        strict_mode = False

    results = []

    for _, row in df.iterrows():
        h, r, t = row["subject"].lower(), row["relation"].lower(), row["object"].lower()

        if strict_mode:
            try:
                h_id = entity_to_id[h]
                r_id = relation_to_id[r]
                h_emb = model.entity_representations[0](torch.tensor([h_id]))
                r_emb = model.relation_representations[0](torch.tensor([r_id]))
                candidate_vec = h_emb + r_emb
                sim = F.cosine_similarity(query_vec, candidate_vec).item()
                results.append((h, r, t, row["image_id"], row["caption"], sim))
            except KeyError:
                continue
        else:
            # Nếu khớp ít nhất 1 thành phần thì thêm vào
            if (query_h in h) or (query_r in r) or (query_t in t):
                results.append((h, r, t, row["image_id"], row["caption"], 0.0))  # không tính sim

    results.sort(key=lambda x: x[-1], reverse=True)
    return results[:top_k]
```
